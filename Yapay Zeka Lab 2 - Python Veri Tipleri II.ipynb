## **Veri Tipleri II: Boolean (True / False)**


Boolean veri tipi yalnızca iki değer alır:
- True
- False

````python
a = True
b = False

print(a)
print(b)
print(type(a))
````

### **1. Karşılaştırma Operatörleri**

| Operatör | Açıklama | Örnek | Sonuç |
|----------|----------|--------|--------|
| `==` | Eşit mi? | `5 == 5` | `True` |
| `!=` | Eşit değil mi? | `5 != 3` | `True` |
| `>`  | Büyük mü? | `10 > 3` | `True` |
| `<`  | Küçük mü? | `2 < 5` | `True` |
| `>=` | Büyük veya eşit mi? | `4 >= 4` | `True` |
| `<=` | Küçük veya eşit mi? | `3 <= 2` | `False` |


````python
print(5 == 5)
print(5 == 3)
print(5 != 3)
````

````python
print(10 > 3)
print(2 != 2)
print(4 <= 6)
````

### **2. String + Boolean**

````python
print("Python" == "Python")
print("Python" == "python")
````

````python
metin = "Python programlama"
print("Python" in metin)
print("Java" in metin)
````

````python
print(len("Python") == 6)
print("a" in "merhaba")
````

### **3. Mantıksal Operatörler (and, or, not)**

#### and

````python
print(5 > 3 and 2 < 4)
print(5 > 3 and 2 > 4)
````

#### or

````python
print(5 > 3 or 2 > 4)
print(1 > 3 or 2 > 4)
````

#### not

````python
print(not True)
print(not False)
````

````python
print(bool(0))
print(bool(5))
````

````python
print(bool(""))
print(bool("Python"))
````
