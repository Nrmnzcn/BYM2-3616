# Fonksiyonlar

## Fonksiyon Nedir?


`Fonksiyon, belirli bir görevi yerine getiren, gerektiğinde çağrılabilen ve isteğe bağlı olarak değer döndüren kod bloklarıdır.`

- Belirli bir işi yaparlar
- Girdi alırlar
- Çıktı üretirler
- Tekrar tekrar kullanılabilirler



## Fonksiyonsuz Kod Problemi

````python
a = 5
b = 7
print(a + b)

x = 10
y = 20
print(x - y)
````

Avantajlar

## Örnek 2: Biyomedikal Alan

* Problem:
Hastanın Vücut Kitle İndeksi (BMI) hesaplama

Formül:
`BMI = kilo / (boy^2)`

````python
def bmi_hesapla(kilo, boy):
    return kilo / (boy ** 2)

def bmi_yorumla(bmi):
    if bmi < 18.5:
        return "Zayıf"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Fazla Kilolu"
    else:
        return "Obez"
````

````python
bmi_nermin = bmi_hesapla(70, 1.68)
bmi_yorumla(bmi_nermin)
````

````python
bmi = bmi_hesapla(89, 1.62)
bmi_yorumla(bmi)
````

## Örnek 3: Döngü + Fonksiyon

İlk **N** sayının toplamı hesaplayabileceğimiz bir fonksiyon yazınız.

````python
def ilk_n_toplam(n):
    toplam = 0
    for i in range(1, n+1):
        toplam += i
    return toplam
````

````python
print(ilk_n_toplam(9))
````

````python
print(ilk_n_toplam(156))
````
