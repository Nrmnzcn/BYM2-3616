## **`break` ve `continue`** Ä°fadeleri

- `break`: dÃ¶ngÃ¼yÃ¼ **tamamen sonlandÄ±rÄ±r**
- `continue`: o turdaki iÅŸlemleri atlar, bir sonraki tura geÃ§er


````python
# break Ã¶rneÄŸi: i 5 olunca dur
for i in range(10):
    if i == 5:
        break
    print(i)
````

````python
# continue Ã¶rneÄŸi: i 5 iken yazdÄ±rma, atla
for i in range(10):
    if i == 5:
        continue
    print(i)
````
