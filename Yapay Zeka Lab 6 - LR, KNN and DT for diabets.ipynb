```python
from sklearn.datasets import load_diabetes
```

```python
# Tüm veri
dataset = load_diabetes()

X = dataset.data # Feature, Özellik
y = dataset.target # Tahmin etmek istediklerim
```

```python
# Özellik -- Xtrain, Xtest
# Target  -- ytain , ytest

from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3)
```

```python
# Bölünmüş veri setlerinin boyutlarını kontrol edelim
print(f"X_train boyutu: {Xtrain.shape}")
print(f"X_test boyutu: {Xtest.shape}")
print(f"y_train boyutu: {ytrain.shape}")
print(f"y_test boyutu: {ytest.shape}")
```

```python
from sklearn import tree
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression


tree_model = tree.DecisionTreeRegressor()
tree_model.fit(Xtrain, ytrain)

knn_model = KNeighborsRegressor()
knn_model.fit(Xtrain, ytrain)

lr_model = LinearRegression()
lr_model.fit(Xtrain, ytrain)
```

```python
# tahmin
y_pred_tree = tree_model.predict(Xtest)
y_pred_knn = knn_model.predict(Xtest)
y_pred_lr = lr_model.predict(Xtest)
```

```python
# hata hesaplama
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error

mae_tree = mean_absolute_error(ytest, y_pred_tree)
mse_tree = mean_squared_error(ytest, y_pred_tree)
rmse_tree = root_mean_squared_error(ytest, y_pred_tree)

mae_knn = mean_absolute_error(ytest, y_pred_knn)
mse_knn = mean_squared_error(ytest, y_pred_knn)
rmse_knn = root_mean_squared_error(ytest, y_pred_knn)

mae_lr = mean_absolute_error(ytest, y_pred_lr)
mse_lr = mean_squared_error(ytest, y_pred_lr)
rmse_lr = root_mean_squared_error(ytest, y_pred_lr)
```

```python
# 5. Sonuçları yazdır
print("Tree Modeli Performansı:")
print(f"  MAE : {mae_tree:.3f}")
print(f"  MSE : {mse_tree:.3f}")
print(f"  RMSE: {rmse_tree:.3f}")

print("KNN Modeli Performansı:")
print(f"  MAE : {mae_knn:.3f}")
print(f"  MSE : {mse_knn:.3f}")
print(f"  RMSE: {rmse_knn:.3f}")

print("LR Modeli Performansı:")
print(f"  MAE : {mae_lr:.3f}")
print(f"  MSE : {mse_lr:.3f}")
print(f"  RMSE: {rmse_lr:.3f}")
```
